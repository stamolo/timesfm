```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
# Конфигурация
INPUT_LENGTH = 100    # Длина входной последовательности
FORECAST_STEPS = 20   # Количество шагов прогнозирования
BATCH_SIZE = 64       # Размер батча
NUM_EPOCHS = 10       # Количество эпох обучения
```

```python
# Генерация синтетических данных с визуализацией
def generate_and_visualize_data(num_samples=10000,
                               input_length=INPUT_LENGTH,
                               forecast_steps=FORECAST_STEPS):
    np.random.seed(42)
    time = np.linspace(0, 20, num_samples + input_length + forecast_steps)

    # Создаем комплексный сигнал
    signal = (
        5.0 *time* np.sin(60 * time) +
        0.6 * np.cos(2.0 * time) +
        0.5 * np.sin(1.5 * time + 1.0) +
        0.3 * np.random.randn(len(time))
    )

    # Визуализация сырых данных
    plt.figure(figsize=(12, 4))
    plt.plot(time[:500], signal[:500], label='Исходный сигнал')
    plt.title("Пример сгенерированных данных (первые 500 точек)")
    plt.xlabel("Время")
    plt.ylabel("Значение")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Нормализация
    scaler = MinMaxScaler(feature_range=(-1, 1))
    signal = scaler.fit_transform(signal.reshape(-1, 1)).flatten()

    # Формирование последовательностей
    X, y = [], []
    for i in range(len(signal) - input_length - forecast_steps + 1):
        X.append(signal[i:i+input_length])
        y.append(signal[i+input_length:i+input_length+forecast_steps])

    return np.array(X), np.array(y), scaler

    # Модифицированная модель для многократного прогнозирования
class ForecastingModel(nn.Module):
    def __init__(self,
                 input_length=INPUT_LENGTH,
                 forecast_steps=FORECAST_STEPS,
                 embed_size=64,
                 **kwargs):
        super().__init__()

        # Параметры модели
        self.input_length = input_length
        self.forecast_steps = forecast_steps

        # Энкодер
        self.input_embedding = nn.Conv1d(1, embed_size, kernel_size=5, padding=2)
        self.position_encoder = PositionalEncoding(embed_size, max_len=input_length)
        self.transformer = TransformerEncoderLayer(
            d_model=embed_size,
            nhead=8,
            dim_feedforward=512,
            batch_first=True
        )

        # Декодер прогнозирования
        self.decoder = nn.Sequential(
            nn.Linear(input_length * embed_size, 512),
            nn.ReLU(),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Linear(256, forecast_steps)
        )

    def forward(self, x):
        # Кодирование
        x = x.permute(0, 2, 1)
        x = self.input_embedding(x)
        x = x.permute(0, 2, 1)
        x = self.position_encoder(x)
        x = self.transformer(x)

        # Прогнозирование
        x = x.reshape(x.size(0), -1)
        return self.decoder(x)

# Генерация и загрузка данных
X, y, scaler = generate_and_visualize_data()
X_tensor = torch.FloatTensor(X).unsqueeze(-1).to(device)
y_tensor = torch.FloatTensor(y).to(device)

# Разделение данных
train_size = int(0.8 * len(X))
train_dataset = torch.utils.data.TensorDataset(X_tensor[:train_size], y_tensor[:train_size])
test_dataset = torch.utils.data.TensorDataset(X_tensor[train_size:], y_tensor[train_size:])

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

# Инициализация модели
model = ForecastingModel(
    input_length=INPUT_LENGTH,
    forecast_steps=FORECAST_STEPS,
    embed_size=64
).to(device)

# Обучение
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-5)
criterion = nn.MSELoss()
```

```python
# Визуализация прогнозов
def plot_predictions(model, dataset, scaler, num_examples=3):
    model.eval()
    plt.figure(figsize=(15, 3*num_examples))

    for idx in range(num_examples):
        with torch.no_grad():
            inputs, target = dataset[idx]
            prediction = model(inputs.unsqueeze(0)).squeeze()

        # Денормализация
        inputs = scaler.inverse_transform(inputs.cpu().numpy()).flatten()
        target = scaler.inverse_transform(target.cpu().numpy().reshape(-1, 1)).flatten()
        prediction = scaler.inverse_transform(prediction.cpu().numpy().reshape(-1, 1)).flatten()

        # Построение графика
        plt.subplot(num_examples, 1, idx+1)

        # Входные данные
        plt.plot(range(INPUT_LENGTH), inputs, 'b-', label='Входная последовательность')

        # Реальные значения
        plt.plot(range(INPUT_LENGTH, INPUT_LENGTH+FORECAST_STEPS), target,
                'go-', markersize=5, label='Реальные значения')

        # Прогноз
        plt.plot(range(INPUT_LENGTH, INPUT_LENGTH+FORECAST_STEPS), prediction,
                'rx--', markersize=6, linewidth=1.5, label='Прогноз')

        plt.axvline(INPUT_LENGTH-0.5, color='gray', linestyle='--')
        plt.ylabel('Значение')
        if idx == 0:
            plt.title('Прогнозирование временного ряда')
            plt.legend()
        if idx == num_examples-1:
            plt.xlabel('Временные шаги')

    plt.tight_layout()
    plt.show()

# Вывод результатов
plot_predictions(model, test_dataset, scaler, num_examples=3)
```
